```python
import requests, re, os, json
from bs4 import BeautifulSoup

def get_products_num(l):
    if l[:5] != "https":
        l = url + l[1:]
    return l
```

```python
headers = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36"
}

url = "https://www.aboutyou.de/"
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text)
```

```python
d = {}
for l in soup.find_all("a"):
    if l and l.get("href") and l.get("href")[:3] == "/fr" and len(l.get("href").split("/")) > 2:
        r = l.get("href").split("/")
        if r[-1] and r[-1] not in d and l.string:
            d[r[-1]] = l.string
            
def trans(s):
    if s in d:
        return d[s]
    return s
```

```python
contents = {}
for l in soup.find_all("a"):
    if l and l.get("href") and l.get("href")[:3] == "/fr" and len(l.get("href").split("/")) > 2:
#         print(l, get_products_num(l.get("href")))
        r = l.get("href").split("/")
        r = [trans(x) for x in r]
        if len(r) > 2 and (r[2] not in contents) and len(r[2])<20:
            contents[r[2]] = {}
        if len(r) > 3 and (r[3] not in contents[r[2]]):
            contents[r[2]][r[3]] = {"":get_products_num(l.get("href"))}
        if len(r) > 4 and (r[4] not in contents[r[2]][r[3]]) and len(r[4])<15:
            try:
                contents[r[2]][r[3]][r[4]] = get_products_num(l.get("href"))
            except:
                contents[r[2]][r[3]] = {}
                contents[r[2]][r[3]][r[4]] = get_products_num(l.get("href"))
```
